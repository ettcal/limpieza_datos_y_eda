### 20. Observa los ratings que te interesen y ahora filtra el dataframe con ese rating para ver cuáles son las películas con dicho rating

### 21. Obten los valores únicos de la variable Rating y la frecuencia total de cada uno de estos valores. Posteriormente crea un nuevo dataframe con esos valores

```python
valores_unicos = df['Rating'].unique()
frecuencias = df['Rating'].value_counts()


nuevo_df = pd.DataFrame({
    'Rating': frecuencias.index,      # valores únicos ordenados por frecuencia
    'Frecuencia': frecuencias.values  # cantidad de cada rating
})
```

### 22. Ordena el nuevo dataframe por la variable `rating`

```python
nuevo_df = nuevo_df.sort_values(by='Rating')
```

### 23. Crea una gráfica de barras con este nuevo dataframe ordenado

```python
plt.figure(figsize=(8, 5))
plt.bar(nuevo_df['Rating'], nuevo_df['Frecuencia'],edgecolor='black')

plt.title('Frecuencia de cada Rating')
plt.xlabel('Rating')
plt.ylabel('Frecuencia')
```

### 24. Crea la matriz de correlación del dataframe de películas

```python
import seaborn as sns

df_corr = df[['Year', 'Runtime_Minutes', 'Rating', 'Votes', 'Revenue_Millions', 'Metascore']].corr()
df_corr

plt.figure(figsize=(8, 6))
sns.heatmap(df_corr, annot=True, cmap='RdBu_r', center=0)
plt.title('Matriz de correlación del DataFrame de películas')


```
